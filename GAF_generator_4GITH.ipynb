```python
# Importing the necessary libraries

# Mathematical operations
import math
import numpy as np

# Plotting
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# This is a simple example of creating a picture from a time series, using a Gramian angular field
# Creating a figure for image output
plt.figure(figsize=(3, 3))
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)

# Specify the size and create the data
size = 30
support = np.arange(0, size)

# Calculate the cosines
time_serie = np.cos(support)

# Creating a GAF

# Performing Min-Max standardization
min_ = np.amin(time_serie)
max_ = np.amax(time_serie)
scaled_serie = (2*time_serie - max_ - min_)/(max_ - min_)

# Values exceeding 1 and -1 are equated to 1 and -1
scaled_serie = np.where(scaled_serie >= 1., 1., scaled_serie)
scaled_serie = np.where(scaled_serie <= -1., -1., scaled_serie)

# We get cosines (these are our polar coordinates)
phi = np.arccos(scaled_serie)

# Creating a coordinate matrix of coordinates from vectors (in fact, all combinations of phi-values)
a, b = np.meshgrid(phi, phi, sparse=True)

# Adding up the coordinates
c = a + b

# Calculating the GAF
# vectorize returns the vectorized version of the function. The argument is passed separately
gaf = np.vectorize(math.cos)(c)

# GAF Image
plt.axis('off')
plt.imshow(gaf, cmap='viridis', interpolation='nearest')
plt.show()
```
